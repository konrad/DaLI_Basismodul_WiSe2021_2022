```python
# Weiter Packages in das laden

import urllib.request
import pandas as pd
```

```python

# https://github.com/konrad/DaLI_Basismodul_WiSe2021_2022

# Die Daten kommen ursprünglich von
# https://www.umweltbundesamt.de/daten/luft/luftdaten/luftqualitaet
# https://meteostat.net/de/place/45I0JU?t=2020-11-01/2021-10-31

air_quality_url = "https://raw.githubusercontent.com/konrad/DaLI_Basismodul_WiSe2021_2022/main/data/air_qualities_cologne.csv"
weather_url = "https://raw.githubusercontent.com/konrad/DaLI_Basismodul_WiSe2021_2022/main/data/weather_cologne.csv"

air_quality_file = "air_quality_cologne.csv"
weather_file = "weather_cologne.csv"
```

```python
urllib.request.urlretrieve(air_quality_url, air_quality_file)
```

```python
urllib.request.urlretrieve(weather_url, weather_file)
```

```python
weather = pd.read_csv(weather_file)
```

```python
air_quality = pd.read_csv(air_quality_file, na_values=["-"])
```

```python
air_quality_and_weather = air_quality.merge(weather, left_on="Datum", right_on="date")
```

```python
air_quality_and_weather
```

```python
air_quality_and_weather.hist("Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³")
```

```python
air_quality_and_weather.hist("Feinstaub (PM₁₀) stündlich gleitendes Tagesmittel in µg/m³", bins=30, color="pink")
```

```python
air_quality_and_weather_selection = air_quality_and_weather[
    ["Stationscode", "Datum", "Uhrzeit", "Luftqualitätsindex"]]
```
